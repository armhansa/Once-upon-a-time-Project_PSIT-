## 1. แยกข้อมูล

```python
import csv
import string
char = string.ascii_uppercase
yy, mm, dd = 2015, 1, 5
step_time = 'set-history_EOD_'+'%d-%d-%d.csv' % (yy, mm, dd)
file = open('set-history_EOD_2015-01-05.csv', 'r')
temporary = csv.reader(file)
data = [row for row in temporary if len(row[0]) <= 6 and row[0][-1] in char and not '$' in row[0] and not '-' in row[0] and not '!' in row[0]]
```

```python
data
```

```python
for key, group in groupby(data, lambda x: x[1]):
    total = 0
    for item in group:
        total += int(item[2])
    print(item[1], total)
```
